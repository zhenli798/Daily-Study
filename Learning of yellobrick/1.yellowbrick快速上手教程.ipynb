# 1. 读入数据

```python
import pandas as pd
from yellowbrick.datasets import load_bikeshare

X, y = load_bikeshare()
print(X.head())
```

# 2. 使用Rank Features visualizer 去计算每一个特征对之间的皮尔森相关系数


```python
from yellowbrick.features import Rank2D
visualizer = Rank2D(algorithm="pearson")
visualizer.fit_transform(X)
visualizer.show(outpath="test.png")
```

# 3. 使用Direct Data Visualization(JointPlotVisualizer) 进一步观察相关性比较强的特征对内部的关系
该可视化工具绘制了y轴上表观温度和x轴上实际测量温度的散点图，并使用简单线性回归绘制了最佳拟合线。此外，对于温度，单变量分布显示为x轴上方的直方图，对于感觉，显示为y轴旁边的直方图。JointPlotVisualizer提供了特征之间非常强的正相关性以及每个特征的范围和分布的概览。请注意，轴被标准化为0和1之间的空间，这是机器学习中的一种常用技术，用于减少一个特征对另一个特征的影响。
接着我们要从图中删去一些可能由于输入错误导致的一些异常值。
可以看到，更极端的温度会对感知温度产生夸大的影响；天气越冷，人们就越可能相信它是寒冷的，天气越热，人们就越觉得它是温暖的，一般来说，温和的温度对个人的舒适感几乎没有影响。这给了我们一个线索，feelslike可能是一个比temp更好的特性，它承诺了一个更稳定的数据集，同时也减少了遇到异常值或错误的风险。
我们可以通过对我们的模型进行任何一个值的训练，并对结果进行评分，最终确认这一假设。如果temp值确实不太可靠，我们应该删除temp变量以支持feelslike。同时，由于没有异常值和输入错误，我们将使用feelslike值。

```python
from yellowbrick.features import JointPlotVisualizer

visualizer = JointPlotVisualizer(columns=['temp', 'feelslike'])
visualizer.fit_transform(X, y)
visualizer.show()
```

接着，我们可以训练我们的模型；让我们将线性回归拟合到模型中，并绘制残差图。

```python
from yellowbrick.regressor import ResidualsPlot
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1
)

visualizer = ResidualsPlot(LinearRegression())
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

发现评分较低，可能使用线性回归模型并不是很适合，于是想着使用正则化去拟合更好的模型，这时候顺便用一下新的可视化工具

```python
import numpy as np

from sklearn.linear_model import RidgeCV
from yellowbrick.regressor import AlphaSelection

alphas = np.logspace(-10, 1, 200)
visualizer = AlphaSelection(RidgeCV(alphas=alphas))
visualizer.fit(X, y)
visualizer.show()
```

找到了最合适的alpha值，接着我们可以训练最终模型并使用PredictionError visualizer将其可视化

```python
from sklearn.linear_model import Ridge
from yellowbrick.regressor import PredictionError

visualizer = PredictionError(Ridge(alpha=3.181))
visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

预测误差可视化工具将实际（测量）值与预期（预测）值相互绘制。黑色虚线是表示零误差的45度线。和残差图一样，这允许我们看到误差发生的位置以及误差的大小。

在这个图中，我们可以看到大多数实例密度小于200个车手。我们可能希望尝试正交匹配追踪或样条曲线来拟合考虑更多区域性的回归。我们还可以注意到，残差图中奇怪的拓扑结构似乎是使用岭回归固定的，并且我们的模型在大值和小值之间有更多的平衡。脊正则化可能解决了我们在两个特征之间的协方差问题。当我们使用其他模型形式进行分析时，我们可以继续使用可视化工具快速比较和查看结果。
